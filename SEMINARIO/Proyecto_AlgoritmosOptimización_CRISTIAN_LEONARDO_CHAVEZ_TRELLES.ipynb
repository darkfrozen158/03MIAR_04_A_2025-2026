# **9.IMPLEMENTACI√ìN DE ALGORITMO VORAZ:**

Un algoritmo voraz (tambi√©n llamado "algoritmo goloso" o "greedy algorithm" en ingl√©s) es un paradigma algor√≠tmico que sigue la heur√≠stica de hacer la elecci√≥n √≥ptima local en cada etapa con la esperanza de encontrar un √≥ptimo global.

Se utiliza un algoritmo voraz porque el problema

‚úÖ Requiere asignar tomas a d√≠as respetando la capacidad m√°xima (6 tomas por d√≠a).

‚úÖ Busca minimizar el costo total, calculado como la suma de actores que asisten cada d√≠a.

‚úÖ Es un problema NP-hard de particionamiento y scheduling con restricciones y minimizaci√≥n de costos combinados.

Se tiene como ventaja:

‚úîÔ∏è Sencillez y facilidad de implementaci√≥n.

‚úîÔ∏è Generalmente m√°s r√°pidos que algoritmos exhaustivos o de programaci√≥n din√°mica.

‚úîÔ∏è Adecuados cuando la soluci√≥n √≥ptima puede construirse a partir de √≥ptimos locales.





```
def cargar_datos(archivo):
    """Carga los datos del archivo CSV y los prepara para el procesamiento"""
    df = pd.read_csv(archivo)

    tomas = []
    for _, row in df.iterrows():
        toma = row['TOMA']
        actores = [actor for actor in df.columns[1:] if row[actor] == 1]
        tomas.append({'toma': toma, 'actores': frozenset(actores)})

    return tomas

def planificacion_voraz(tomas, max_tomas_por_dia=6):
    """
    Implementaci√≥n del algoritmo voraz para planificar las sesiones de doblaje

    Variables:
        tomas: Lista de diccionarios con las tomas y sus actores
        max_tomas_por_dia: M√°ximo de tomas que se pueden grabar en un d√≠a

    Resultado:
        dias: Lista con la planificaci√≥n por d√≠as
        costo_total: Costo total calculado
        dias_por_actor: D√≠as trabajados por cada actor
    """
    # Ordenar tomas por frecuencia de combinaci√≥n de actores (heur√≠stica)
    frecuencias = defaultdict(int)
    for toma in tomas:
        frecuencias[toma['actores']] += 1

    tomas_ordenadas = sorted(tomas, key=lambda x: (-frecuencias[x['actores']], len(x['actores'])))

    dias = []
    dias_por_actor = defaultdict(set)

    while tomas_ordenadas:
        # Tomar la primera toma no asignada
        toma_actual = tomas_ordenadas.pop(0)
        actores_actuales = toma_actual['actores']

        # Crear nuevo d√≠a
        dia = {
            'tomas': [toma_actual['toma']],
            'actores': actores_actuales
        }

        # Registrar d√≠as trabajados
        dia_num = len(dias) + 1
        for actor in actores_actuales:
            dias_por_actor[actor].add(dia_num)

        # Buscar tomas compatibles para el mismo d√≠a
        i = 0
        while i < len(tomas_ordenadas) and len(dia['tomas']) < max_tomas_por_dia:
            if tomas_ordenadas[i]['actores'] == actores_actuales:
                # Agregar al d√≠a actual
                dia['tomas'].append(tomas_ordenadas.pop(i)['toma'])
            else:
                i += 1

        dias.append(dia)

    # Calcular costo total (suma de d√≠as trabajados por cada actor)
    costo_total = sum(len(dias) for dias in dias_por_actor.values())

    return dias, costo_total, dias_por_actor

def mostrar_resultados(dias, costo_total, dias_por_actor):
    """Muestra los resultados de la planificaci√≥n"""
    print("PLANIFICACION OPTIMA DE SESIONES DE DOBLAJE:")
    for i, dia in enumerate(dias, 1):
        print(f"D√≠a {i}: Tomas {', '.join(map(str, dia['tomas']))} - Actores: {', '.join(dia['actores'])}")

    print(f"\nCOSTO TOTAL (d√≠as de trabajo por actores): {costo_total}")

    print("\nD√çAS TRABAJADOS POR CADA ACTOR:")
    for actor in sorted(dias_por_actor.keys()):
        dias_trabajados = sorted(dias_por_actor[actor])
        print(f"{actor}: {len(dias_trabajados)} d√≠a(s) - D√≠as: {', '.join(map(str, dias_trabajados))}")

    # Estad√≠sticas
    total_dias = len(dias)
    actores_por_dia = [len(dia['actores']) for dia in dias]
    tomas_por_dia = [len(dia['tomas']) for dia in dias]

    print("\nESTADISTICAS GENERALES:")
    print(f"TOTAL DE DIAS DE GRABACI√ìN: {total_dias}")
    print(f"PROMEDIO DE ACTORES POR D√çA: {sum(actores_por_dia)/total_dias:.2f}")
    print(f"PROMEDIO DE TOMAS POR D√çA: {sum(tomas_por_dia)/total_dias:.2f}")
    print(f"TASA DE UTILIZACI√ìN (tomas/d√≠a): {sum(tomas_por_dia)/(total_dias*6)*100:.1f}%")

def main():
    # Cargar datos
    tomas = cargar_datos(url_data)

    # Ejecutar algoritmo voraz
    dias, costo_total, dias_por_actor = planificacion_voraz(tomas)

    # Mostrar resultados
    mostrar_resultados(dias, costo_total, dias_por_actor)

if __name__ == "__main__":
    main()
```

**RESPUESTA:**

Por tanto la cantidad de posibilidades sin restricciones es: n!

Ahora utilizamos los numeros de BELL para tener en cuenta el n√∫mero total de formas posibles de particionar un conjunto de
n elementos en subconjuntos no vac√≠os (sin restricciones).

image.png

Consideraciones :
*   Tomas: 30 tomas distintas (numeradas del 1 al 30).

*   D√≠as: Las tomas se agrupan en d√≠as, donde cada d√≠a es un subconjunto no vac√≠o de tomas.

*   Sin restricciones: No hay l√≠mite en el n√∫mero de tomas por d√≠a ni en la compatibilidad de actores.

**IMPLEMENTACI√ìN EN CODIGO PYTHON:**

```
def numeros_bell(n):
    """Calcula el n√∫mero de Bell B_n usando programaci√≥n din√°mica."""
    if n == 0:
        return 1
    # Inicializar una tabla para almacenar los n√∫meros de Stirling de segunda clase
    S = [[0] * (n + 1) for _ in range(n + 1)]
    S[0][0] = 1  # Caso base

    for i in range(1, n + 1):
        for k in range(1, i + 1):
            S[i][k] = k * S[i - 1][k] + S[i - 1][k - 1]

    # El n√∫mero de Bell es la suma de la fila S[n]
    return sum(S[n])
```

```
print("B = 30, es", numeros_bell(30))
```

**RESPUESTA:**

Ademas el n√∫mero de posibilidades v√°lidas es extremadamente grande pero dif√≠cil de calcular exactamente.

Es una combinaci√≥n de 30 tomas que pueden ser escogidas de 6 tomas en 6 tomas (tomas por d√≠a),
Donde las posibles muestras sin orden de las 6 tomas distintas que se pueden extraer del totas de 30 tomas.

Por tanto la cantidad de posibilidades con restricciones se calcula con la f√≥rmula: n! / (r!*(n-r)!)

Para la pregunta n = 30 y r = 6.  
Aplicando la formula: 30!/(6!*(30-6)!)
El resultado es: 593,775 posibilidades

Dado que el c√°lculo exacto es inviable, usaremos una aproximaci√≥n basada en permutaciones y poda:

**IMPLEMENTACI√ìN EN CODIGO PYTHON:**

```
tomas = []
actores_por_toma = []
for _, row in df.iterrows():
    tomas.append(row['TOMA'])
    actores_por_toma.append([actor for actor in row.index[1:] if row[actor] == 1])

# Construir matriz de compatibilidad
n = len(tomas)
compatible = [[False] * n for _ in range(n)]
for i in range(n):
    for j in range(n):
        compatible[i][j] = len(set(actores_por_toma[i]) & set(actores_por_toma[j])) > 0

# Funci√≥n para verificar si un grupo de tomas es v√°lido (compatible y ‚â§6 tomas)
def es_grupo_valido(grupo):
    for i, j in combinations(grupo, 2):
        if not compatible[i][j]:
            return False
    return len(grupo) <= 6

# Estimaci√≥n del n√∫mero de soluciones (aproximaci√≥n muy simplificada)
# Usamos una permutaci√≥n aleatoria y contamos grupos v√°lidos (Monte Carlo)
def estimar_soluciones(num_muestras=1000):
    conteo = 0
    for _ in range(num_muestras):
        perm = random.sample(range(n), n)
        dias = []
        grupo_actual = []
        for toma in perm:
            grupo_actual.append(toma)
            if not es_grupo_valido(grupo_actual):
                dias.append(grupo_actual[:-1])
                grupo_actual = [toma]
        if grupo_actual:
            dias.append(grupo_actual)
        if all(len(d) <= 6 for d in dias):
            conteo += 1
    return (conteo / num_muestras) * factorial(n)  # Ajuste por todas las permutaciones
```

```

# Ejemplo de estimaci√≥n (muy lento para n=30)
print("Estimaci√≥n del n√∫mero de soluciones v√°lidas:", estimar_soluciones(num_muestras=10))
```

**IMPLEMENTACI√ìN EN CODIGO PYTHON:**

```
# Preprocesamiento: Convertir tomas a conjuntos de actores
tomas = []
for _, row in df.iterrows():
    toma = row['TOMA']
    actores = {actor for actor in df.columns[1:] if row[actor] == 1}
    tomas.append((toma, actores))

MAX_TOMAS_POR_DIA = 6
NUM_TOMAS = len(tomas)

def es_compatible(grupo, nueva_toma):
    """Verifica si una nueva toma es compatible con un grupo existente"""
    if not grupo:
        return True
    # Todas las tomas en el grupo deben tener los mismos actores que la nueva toma
    return all(toma[1] == nueva_toma[1] for toma in grupo)

def calcular_costo(asignacion):
    """Calcula el costo total de una asignaci√≥n"""
    dias_actores = set()
    for dia in asignacion:
        for toma in dia:
            for actor in toma[1]:
                dias_actores.add((actor, asignacion.index(dia)))
    return len({actor for actor, dia in dias_actores})

def fuerza_bruta(tomas, max_por_dia=6):
    """Algoritmo de fuerza bruta para encontrar la soluci√≥n √≥ptima"""
    mejor_asignacion = None
    mejor_costo = float('inf')

    # Generar todas las posibles particiones v√°lidas
    from more_itertools import set_partitions
    from tqdm import tqdm  # Para barra de progreso (opcional)

    for particion in tqdm(set_partitions(range(len(tomas))), desc="Evaluando particiones"):
        # Verificar que ning√∫n grupo exceda el m√°ximo por d√≠a
        valida = all(len(grupo) <= max_por_dia for grupo in particion)
        if not valida:
            continue

        # Convertir √≠ndices a tomas reales
        asignacion = []
        for grupo in particion:
            grupo_tomas = [tomas[i] for i in grupo]
            # Verificar compatibilidad dentro del grupo
            if all(es_compatible(grupo_tomas[:i], grupo_tomas[i]) for i in range(1, len(grupo_tomas))):
                asignacion.append(grupo_tomas)
            else:
                break
        else:
            # Si todas las tomas son compatibles en sus grupos
            costo_actual = calcular_costo(asignacion)
            if costo_actual < mejor_costo:
                mejor_costo = costo_actual
                mejor_asignacion = asignacion

    return mejor_asignacion, mejor_costo

# Debido a la complejidad computacional, primero probaremos con un subconjunto peque√±o
print("Ejecutando algoritmo de fuerza bruta")
tomas_prueba = tomas[:10]  # Probamos con las primeras 10 tomas para demostraci√≥n
asignacion_optima, costo_optimo = fuerza_bruta(tomas_prueba)

# Mostrar resultados
print("\nMejor asignaci√≥n encontrada:")
for i, dia in enumerate(asignacion_optima, 1):
    tomas_dia = [t[0] for t in dia]
    actores_dia = next(iter(dia))[1] if dia else set()
    print(f"D√≠a {i}: Tomas {', '.join(map(str, tomas_dia))} - Actores: {', '.join(actores_dia)}")

print(f"\nCosto total √≥ptimo: {costo_optimo}")

# Calcular d√≠as trabajados por actor
dias_actores = defaultdict(set)
for dia_idx, dia in enumerate(asignacion_optima, 1):
    for toma in dia:
        for actor in toma[1]:
            dias_actores[actor].add(dia_idx)

print("\nD√≠as trabajados por cada actor:")
for actor in sorted(dias_actores.keys()):
    print(f"{actor}: {len(dias_actores[actor])} d√≠a(s) - D√≠as: {', '.join(map(str, sorted(dias_actores[actor])))}")
```

**Dise√±a un algoritmo que mejore la complejidad del algortimo por fuerza bruta.**

**RESPUESTA:**

 El algoritmo a implementar es utilizando la tecnica de busqueda local es una t√©cnica de optimizaci√≥n heur√≠stica que inicia con una soluci√≥n inicial factible e iterativamente explora soluciones cercanas (vecindario) para encontrar una mejor. Y Se detiene cuando ninguna soluci√≥n vecina es mejor que la actual (√≥ptimo local).

**ARGUMENTACI√ìN PORQUE EL ALGORITMO DE BUSQUEDA LOCAL ES MEJOR QUE EL ALGORITMO DE FUERZA BRUTA:**

*  Porque el algoritmo B√∫squeda Local forma parte de una soluci√≥n inicial
factible y se encarga de explorar
soluciones cercanas (vecindario) para encontrar mejoras y se detiene en un √≥ptimo local, no necesariamente global. Es muy util y mucho m√°s r√°pido problemas grandes.

*  Ademas que el tiempo de ejecuci√≥n de los codigos es en Segundos o minutos, dependiendo de la cantidad de interaciones.

*  Uso practico, debido a que empresas y estudios necesitan respuestas r√°pidas con soluciones suficientemente buenas, no la soluci√≥n te√≥rica perfecta si tarda d√≠as en calcularse.

*  Es ideal porque es pr√°ctico y eficiente para problemas reales de tama√±o moderado a grande.

```
# Preprocesamiento: Convertir a lista de tomas con sus actores
tomas = []
for _, row in df.iterrows():
    toma = row['TOMA']
    actores = [actor for actor in df.columns[1:] if row[actor] == 1]
    tomas.append({'toma': toma, 'actores': frozenset(actores)})

MAX_TOMAS_POR_DIA = 6

## 1. Funci√≥n para generar soluci√≥n inicial (usando el voraz mejorado)
def solucion_inicial_voraz(tomas):
    # Agrupar tomas por combinaci√≥n de actores
    grupos = defaultdict(list)
    for toma in tomas:
        grupos[toma['actores']].append(toma['toma'])

    # Crear d√≠as llenando con tomas del mismo grupo
    dias = []
    for actores, tomas_grupo in grupos.items():
        # Dividir en chunks de m√°ximo MAX_TOMAS_POR_DIA
        for i in range(0, len(tomas_grupo), MAX_TOMAS_POR_DIA):
            chunk = tomas_grupo[i:i+MAX_TOMAS_POR_DIA]
            dias.append({'tomas': chunk, 'actores': actores})

    return dias

## 2. Funci√≥n para calcular el costo de una soluci√≥n
def calcular_costo(dias):
    actores_dias = defaultdict(set)
    for dia_idx, dia in enumerate(dias, 1):
        for actor in dia['actores']:
            actores_dias[actor].add(dia_idx)
    return sum(len(dias) for dias in actores_dias.values())

## 3. Generar vecinos (operadores de b√∫squeda local)
def generar_vecino(dias):
    # Hacer una copia profunda
    vecino = [{'tomas': list(dia['tomas']), 'actores': dia['actores']} for dia in dias]

    # Elegir un operador aleatorio
    operador = random.choice([
        'mover_toma',
        'intercambiar_tomas',
        'dividir_dia',
        'combinar_dias_compatibles'
    ])

    try:
        if operador == 'mover_toma' and len(vecino) > 1:
            # Seleccionar una toma aleatoria y moverla a un d√≠a compatible
            dia_origen_idx = random.randint(0, len(vecino)-1)
            if not vecino[dia_origen_idx]['tomas']:
                return vecino

            toma_idx = random.randint(0, len(vecino[dia_origen_idx]['tomas'])-1)
            toma = vecino[dia_origen_idx]['tomas'].pop(toma_idx)

            # Encontrar d√≠as compatibles (mismos actores y espacio)
            for dia_destino_idx, dia in enumerate(vecino):
                if (dia['actores'] == tomas[toma-1]['actores'] and
                    len(dia['tomas']) < MAX_TOMAS_POR_DIA and
                    dia_destino_idx != dia_origen_idx):
                    dia['tomas'].append(toma)
                    break
            else:
                # Si no hay d√≠a compatible, crear uno nuevo
                vecino.append({
                    'tomas': [toma],
                    'actores': tomas[toma-1]['actores']
                })

            # Eliminar d√≠a origen si qued√≥ vac√≠o
            if not vecino[dia_origen_idx]['tomas']:
                vecino.pop(dia_origen_idx)

        elif operador == 'intercambiar_tomas' and len(vecino) > 1:
            # Seleccionar dos tomas de diferentes d√≠as e intercambiarlas si es posible
            dia1_idx, dia2_idx = random.sample(range(len(vecino)), 2)
            if not vecino[dia1_idx]['tomas'] or not vecino[dia2_idx]['tomas']:
                return vecino

            toma1_idx = random.randint(0, len(vecino[dia1_idx]['tomas'])-1)
            toma2_idx = random.randint(0, len(vecino[dia2_idx]['tomas'])-1)

            toma1 = vecino[dia1_idx]['tomas'][toma1_idx]
            toma2 = vecino[dia2_idx]['tomas'][toma2_idx]

            # Verificar compatibilidad
            if (tomas[toma1-1]['actores'] == vecino[dia2_idx]['actores'] and
                tomas[toma2-1]['actores'] == vecino[dia1_idx]['actores']):
                vecino[dia1_idx]['tomas'][toma1_idx] = toma2
                vecino[dia2_idx]['tomas'][toma2_idx] = toma1

        elif operador == 'dividir_dia':
            # Seleccionar un d√≠a con muchas tomas y dividirlo
            dias_largos = [i for i, dia in enumerate(vecino) if len(dia['tomas']) > 1]
            if not dias_largos:
                return vecino

            dia_idx = random.choice(dias_largos)
            dia = vecino.pop(dia_idx)
            mitad = len(dia['tomas']) // 2
            vecino.extend([
                {'tomas': dia['tomas'][:mitad], 'actores': dia['actores']},
                {'tomas': dia['tomas'][mitad:], 'actores': dia['actores']}
            ])

        elif operador == 'combinar_dias_compatibles' and len(vecino) > 1:
            # Intentar combinar dos d√≠as con los mismos actores
            compatibles = []
            for i in range(len(vecino)):
                for j in range(i+1, len(vecino)):
                    if (vecino[i]['actores'] == vecino[j]['actores'] and
                        len(vecino[i]['tomas']) + len(vecino[j]['tomas']) <= MAX_TOMAS_POR_DIA):
                        compatibles.append((i, j))

            if compatibles:
                i, j = random.choice(compatibles)
                if i < j:
                    j -= 1  # Ajustar √≠ndice despu√©s de pop
                vecino[i]['tomas'].extend(vecino.pop(j)['tomas'])

    except:
        pass  # Si algo falla, devolver el vecino sin cambios

    return vecino

## 4. Algoritmo de b√∫squeda local
def busqueda_local(tomas, max_iter=1000, max_sin_mejora=100):
    # Generar soluci√≥n inicial
    mejor_sol = solucion_inicial_voraz(tomas)
    mejor_costo = calcular_costo(mejor_sol)

    sin_mejora = 0
    for iteracion in range(max_iter):
        if sin_mejora >= max_sin_mejora:
            break

        # Generar vecino
        vecino = generar_vecino(mejor_sol)
        costo_vecino = calcular_costo(vecino)

        # Evaluar si es mejor
        if costo_vecino < mejor_costo:
            mejor_sol = vecino
            mejor_costo = costo_vecino
            sin_mejora = 0
        else:
            sin_mejora += 1

    return mejor_sol, mejor_costo

# Ejecutar la b√∫squeda local
solucion_optima, costo_optimo = busqueda_local(tomas)

# Mostrar resultados
print("\nResultados de ejecuci√≥n por Algoritmo de Busqueda Local:")
for i, dia in enumerate(solucion_optima, 1):
    print(f"D√≠a {i}: Tomas {', '.join(map(str, dia['tomas']))} - Actores: {', '.join(dia['actores'])}")

print(f"\nCosto total √≥ptimo: {costo_optimo}")

# Calcular d√≠as trabajados por actor
dias_actores = defaultdict(set)
for dia_idx, dia in enumerate(solucion_optima, 1):
    for actor in dia['actores']:
        dias_actores[actor].add(dia_idx)

print("\nD√≠as trabajados por cada actor:")
for actor in sorted(dias_actores.keys()):
    print(f"{actor}: {len(dias_actores[actor])} d√≠a(s) - D√≠as: {', '.join(map(str, sorted(dias_actores[actor])))}")

# Estad√≠sticas
print("\nEstad√≠sticas de la soluci√≥n:")
print(f"Total d√≠as de grabaci√≥n: {len(solucion_optima)}")
print(f"Promedio de tomas por d√≠a: {sum(len(dia['tomas']) for dia in solucion_optima)/len(solucion_optima):.2f}")
```
